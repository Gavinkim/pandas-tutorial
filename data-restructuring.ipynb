```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#pivot
df = pd.DataFrame({
    'foo': ['one','one','one','two','two','two'],
    'bar': ['A','B','C','A','B','C'],
    'baz': [1,2,3,4,5,6]
})
df
```

```python
#pivot
#중복되는 인덱스와 컬럼이 있으면 제거 후 피봇하도록 하자
# df.drop_duplicates(['foo','bar'])
df.pivot(index='foo',columns='bar',values='baz')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(3)
```

```python
titanic.count()
```

```python
#group by survived 로 그룹핑 하고, 그룹핑된 값의 평균값을 출력
titanic_s = titanic.groupby('survived').mean()
titanic_s
```

```python
#그룹핑 된 df 를 피봇팅 하여 새로운 데이터프레임 생성
#행 -> plcass, 열 -> age, 값 -> fare
titanic_s.pivot(index='pclass',columns='age', values='fare')
```

```python
titanic.pivot_table(values='survived',index='sex',columns='class',aggfunc='sum')
```
